```python
# import dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# data collection and processing
```

```python
# loading cdv data to a pandas dataframe
heart_data = pd.read_csv("HeartDisease.csv")
```

```python
# print first 5 rows
heart_data.head()
```

```python
# last five rows of dataset
heart_data.tail()
```

```python
# check number of rows and columns in the dataset
heart_data.shape
```

```python
# getting more info on the dataset
heart_data.info()
```

```python
# checking for missing values
missing_values = heart_data.isnull().sum()
# print(missing_values)
heart_data.replace('', np.nan, inplace=True)  # Replace empty strings with NaN
heart_data.replace('None', np.nan, inplace=True)  # Replace 'None' with NaN if applicable
```

```python
heart_data.head()
```

```python
heart_data.tail()
# i have no missing values
```

```python
# statistical measure of the data
heart_data.describe()
```

```python
# checking the distribution of Target Variables
heart_data["target"].value_counts()
```

```python
1--> defective heart
0--> healthy heart
```

```python
# splitting the Features and Targets
X = heart_data.drop(columns="target", axis=1)
Y = heart_data["target"]
print(X)
```

```python
print(Y)
```

```python
# splitting data into taring data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training
```

```python
# logistic models for classification
model = LogisticRegression()
```

```python
# traing the logistic regression model with Training data
model.fit(X_train, Y_train)
```

```python
# model evaluation
# Accuracy Score
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("accuracy on training data:", training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("accuracy on test data:", test_data_accuracy)
```

```python
# Building a predicting system
```
